```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
img=cv2.imread('lena.png')
imgray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret,thresh =cv2.threshold(imgray,127,255,0)

contours,hierarchy =cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
x=cv2.drawContours(img,contours,-1,(0,255,0),3)
cv2_imshow(x)
```

```python
import cv2
from PIL import Image
pictPath=r'haarcascade_frontalface_default.xml'
face_cascade=cv2.CascadeClassifier(pictPath)
img=cv2.imread("multi_persons.jpg")
cv2.rectangle(img,(img.shape[1]-140,img.shape[0]-20),(img.shape[1],img.shape[0]),(0,255,255),-1)
faces=face_cascade.detectMultiScale(img,scaleFactor=1.1,minNeighbors=3,minSize=(20,20))
cv2.putText(img,"Finding "+str(len(faces))+" face",(img.shape[1]-135,img.shape[0]-5),cv2.FONT_HERSHEY_COMPLEX,0.5,(255,0,0),1)
for(x,y,w,h) in faces:
  cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
  cut = img[y:y+h, x:x+w]
  cv2.imwrite('cut'+str(x)+'.jpg', cut)
cv2_imshow(img)
```

```python
import cv2
from PIL import Image

pictPath= r'haarcascade_frontalface_default.xml'
face_cascade = cv2.CascadeClassifier(pictPath)

img = cv2.imread("Andy_3.jpeg")

img1 = Image.open("Andy_3.jpeg")

faces = face_cascade.detectMultiScale(img, scaleFactor=1.1, minNeighbors=2, minSize=(20,20))

cv2.rectangle(img, (img.shape[1]-140, img.shape[0]-20), (img.shape[1], img.shape[0]), (0,255,255), -1)

cv2.putText( img,"Finding "+str(len(faces))+" face",(img.shape[1]-135,img.shape[0]-5),cv2.FONT_HERSHEY_COMPLEX,0.5,(255,0,0),1)

imgCrop = []
count = 8
for(x,y,w,h) in faces:
  cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
  #imgCrop.append(img1.crop((x, y, w, h)))
  tmp = img1.crop((x, y, x+w, y+h))
  imgCrop.append(tmp)
  tmp.save("/content/AndyNew/face"+ str(count) + ".jpg")
  count += 1

for i in imgCrop:
  i.show()
  display(i)

cv2_imshow(img)
```
