# 4.4 dict的資料型別
➢一般稱為「字典」<br>
➢可由{ }或dict( )建立<br>
<font color=red>*必要條件：要有鍵值對(Key：Value)</font><br>
➢內容不重複<br>
➢內容可修改(Mutable)<br>
<font color=red>*但鍵key必須具不可修改屬性</font><br>
➢無序的<br>



*   <font color=#84C1FF>clear( )</font> 刪除字典內的所有元素
*   <font color=#84C1FF>copy( )</font> 傳回字典的拷貝
*   <font color=#84C1FF>dict.fromkeys(seq,value)</font> 以seq中的元素做為鍵，value為值來建立字典

*   <font color=#84C1FF>keys( )</font> 傳回所有的鍵所組成的物件

*   <font color=#84C1FF>values( )</font> 傳回字典所有的值所組成的物件
*   <font color=#84C1FF>pop(key,default)</font> 刪除key 與其對應的值，傳回值為被刪除的值
*   <font color=#84C1FF>popitem( )</font> 隨機取出一個鍵值對，並以序對傳回它們

```python
def add(n):
  sum = 0
  for i in range(1, n + 1):
    sum += i
  return sum
```

```python
def n_digits(num):
  t = 0
  while num >= 10:
    num /= 10
    t += 1
  return t + 1
```

# 6.2 參數的傳遞機制

```python
def square_tup(tpl):
  return tuple([i**2 for i in tpl])
```

```python
square_tup((2, 4, 6))
```

```python
def combine(t1, t2):
  lst1 = list(t1)
  lst2 = list(t2)
  lst1.extend(t2)
  return tuple(lst1)
```

```python
combine((1, 3), (2, 5, 6))
```

# 6.5 遞迴函數

```python
def r_pow(base, n):
  if n > 1:
    return base * r_pow(base, n - 1)
  else:
    return base
r_pow(2, 8)
```

```python
def r_sum(n):
  if n > 1:
    return n * (n - 1) + r_sum(n - 1)
  else:
    return 0
r_sum(5)
```

```python
class calculate():
  def __init__(self,a,b):
    self.n1 = a
    self.n2 = b
  def add(self):
    return self.n1+self.n2
  def gcd(self):
    return math.gcd(self.n1,self.n2)
  def lcm(self):
    return self.n1*self.n2//math.gcd(self.n1,self.n2)
  def power(self):
    return self.n1**self.n2
```

```python
c0 = calculate(2,10)
```

```python
c0.gcd()
```

```python
c0.lcm()
```

# 7.3 類別函數和靜態函數

```python
class Employee:
  payRate=[1,1.2,1.5]
  def __init__(self,base):
    self.basesalary = base
  def salary(self,hr,bonus):
    return self.basesalary*hr*self.payRate[bonus]
  @classmethod
  def set_payRate(cls,new_payRate):
    cls.payRate = new_payRate
```

```python
c1 = Employee(160)
c1.salary(10,1)
```

```python
c1.set_payRate([1,2,5])
c1.salary(10,1)
```

# 7.4 繼承

```python
class Car:
  def __init__(self,color):
    self.color = color
  def show(self):
    print(f'color = {self.color}')
class Truck(Car):
  def __init__(self,dr,ow,co):
    self.doors = dr
    self.owner = ow
    super().__init__(co)
  def show(self):
    print(f'doors = {self.doors}')
    print(f'owner = {self.owner}')
    print(f'color = {self.color}')
```

```python
Car('white').show()
Truck(4,'Tom','black').show()
```

# 7.5 類別的進階認識


*   Python類別裡成員一般都預設為公有(Public)
*   設定私有方式：在成員名稱前加兩個連續底線



```python
class Circle:
  def __init__(self,r=1,c='red'):
    self.__rad = r
    self.color = c
  def __area(self):
    return 3.14*self.__rad**2
  def print_area(self):
    print('area = ',self.__area())
```

```python
c1 = Circle(2,'yellow')
c1.__area()       #私有
```

```python
c1.print_area()
```

```python
class Triangle:
  def __init__(self,base,height):
    self.__base = base
    self.__height = height
  def area(self):
    return 0.5*self.__base*self.__height
  def change(self,base):
    self.__base = base
```

```python
t0 = Triangle(10,5)
t0.area()
```

```python
t0.change(40)
t0.area()
```

# Regular Expression(RegEx)正則表達式

```python
import re
```

```python
str1 = "cat"
str2 = "bird"
string = "dog runs to cat"
print(re.search(str1,string))    #span表示重複的字在12~15，match表示重複的字串
print(re.search(str2,string))    #找不到則輸出None
```

```python
#可以對比多種可能，用[]
ptn1 = r"run"
ptn2 = r"ran"
ptn3 = r"r[au]n"
print(re.search(ptn1, "dog runs to cat"))
print(re.search(ptn2, "dog runs to cat"))
print(re.search(ptn3, "dog runs to cat"))
```

```python
#可以對比更多種可能，A-Z、0-9都可
print(re.search(r"r[a-z]n", "dog runs to cat"))
```



*   <font color=#84C1FF>\d</font> 表所有數字
*   <font color=#84C1FF>\D</font> 為\d的反面，表所有不是數字
*   <font color=#84C1FF>\s</font> 表所有空白字符 <font color=green>例如:[\t\n\r\f\v]</font>
*   <font color=#84C1FF>\S</font> 為\s的反面
*   <font color=#84C1FF>\w</font> 表[a-zA-z0-9_]
*   <font color=#84C1FF>\W</font> 表\w的反面
*   <font color=#84C1FF>\b</font> 空白字串(只有在字串的頭或尾)
*   <font color=#84C1FF>\B</font> 空白字串(不在字串的頭或尾)
*   <font color=#84C1FF>\\(兩個反斜線)</font> \的對比方式
*   <font color=#84C1FF>.</font> 對比除了\n以外的所有
*   <font color=#84C1FF>^</font> 句首
*   <font color=#84C1FF>$</font> 句尾
*   <font color=#84C1FF>?</font> 是否
*   <font color=#84C1FF>*</font> 出現0或多次
*   <font color=#84C1FF>+</font> 出現1或多次
*   <font color=#84C1FF>{n, m}</font> 出現n~m次
*   清單項目


















```python
#\d\D例子
print(re.search(r"r\dn", "run r4n"))
print(re.search(r"r\Dn", "run r4n"))
#\b\B例子
print(re.search(r"\bruns\b", "dog runs to cat"))
print(re.search(r"\B runs \B", "dog  runs  to cat"))
#^$例子
print(re.search(r"^dog", "dog runs to cat"))
print(re.search(r"cat$", "dog runs to cat"))
#?例子
print(re.search(r"Mon(day)?", "Monday"))
print(re.search(r"Mon(day)?", "Mon"))
#*例子
print(re.search(r"ab*", "a"))
print(re.search(r"ab*", "abbbbb"))
#+例子
print(re.search(r"ab+", "a")) #0次對比不到
print(re.search(r"ab+", "abbbbb"))
#{n, m}例子
print(re.search(r"ab{2,10}", "a"))
print(re.search(r"ab{2,10}", "abbbbb"))
```

```python
#多行對比，M為MULTILINE的簡寫
string = """
dog runs to cat.
I run to dog.
"""
print(re.search(r"^I", string)) #因為多行所以I偵測不到為句首
print(re.search(r"^I", string, flags=re.M))
```

```python
match = re.search(r"(\d+), Date: (.+)", "ID: 021523, Date: Feb/12/2017")
print(match.group())
print(match.group(1))
print(match.group(2))
```

```python
# ?P<>來指定名字
match = re.search(r"(?P<id>\d+), Date: (?P<date>.+)", "ID: 021523, Date: Feb/12/2017")
print(match.group('id'))
print(match.group('date'))
```

```python
#findall
print(re.findall(r"r[au]n", "run ran ren"))
#| : or
print(re.findall(r"r(u|a)n", "run ran ren"))
print(re.findall(r"(run|ran)", "run ran ren"))
```
