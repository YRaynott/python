# M2-2 單一筆資料的視覺化

## M2-2-1 繪製折線圖

```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

##M2-2-2 指定圖表的座標軸範圍

```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi)
y = np.sin(x)

plt.ylim(0, 1)
plt.plot(x, y)
plt.show()
```

##M2-2-3 設定圖表標題與兩軸名稱

```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi)
y = np.sin(x)

plt.title('y = sin(x) (0<y<1)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.ylim(0, 1)
plt.plot(x, y)
plt.show()
```

##M2-2-4 在圖表上顯示網格



```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi)
y = np.sin(x)

plt.title('y = sin(x)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.plot(x, y)
plt.show()
```

##M2-2-5 自訂座標軸的刻度及標籤

```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi)
y = np.sin(x)

plt.title('y = sin(x)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

ticks = [0, np.pi * 0.5, np.pi, np.pi * 1.5, np.pi * 2]


labels = ['0°', '90°', '180°', '270°', '360°']
plt.xticks(ticks, labels)

plt.plot(x, y)
plt.show()
```

#M2-3 多筆資料的視覺化 (test1)

##M2-3-1 在同一張圖表繪製多筆資料並指定不同顏色

```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi)
y1 = np.sin(x)
y2 = np.cos(x)

plt.title('sin(x) and cos(x)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

ticks = [0, np.pi * 0.5, np.pi, np.pi * 1.5, np.pi * 2]


labels = ['0°', '90°', '180°', '270°', '360°']
plt.xticks(ticks, labels)

plt.plot(x, y1, color="orange")
plt.plot(x, y2, color="purple")

plt.show()
```

##M2-3-2 設定圖例

```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi)
y1 = np.sin(x)
y2 = np.cos(x)

plt.title('sin(x) and cos(x)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

ticks = [0, np.pi * 0.5, np.pi, np.pi * 1.5, np.pi * 2]


labels = ['0°', '90°', '180°', '270°', '360°']
plt.xticks(ticks, labels)

plt.plot(x, y1, label = "y1 = sin(x)", color="orange")
plt.plot(x, y2, label = "y2 = cos(x)", color="purple")
plt.legend(loc = 'lower left')

plt.show()
```

# M2-4 繪製內含多張子圖的圖表 (test2)

##M2-4-1 設定整張圖表的尺寸

```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi)
y = np.sin(x)

plt.figure(figsize=(6, 8))
plt.plot(x, y)
plt.show()
```

##M2-4-2 在畫布切出子圖區 , 並繪製內容

```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi)
y1 = np.sin(x)
y5 = np.cos(x)
y6 = np.tan(x)

fig = plt.figure(figsize=(8, 6))		#整個圖表大小為 8 x 6 英吋


ax1 = fig.add_subplot(231)
ax1.plot(y1)

ax2 = fig.add_subplot(235)
ax2.plot(y5)

ax3 = fig.add_subplot(236)
ax3.plot(y6)

plt.show()
```

##M2-4-3 調整子圖間距

```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi)
y1 = np.sin(x)
y5 = np.cos(x)
y6 = np.tan(x)

fig = plt.figure(figsize=(8, 6))		#整個圖表大小為 8 x 6 英吋


ax1 = fig.add_subplot(231)
ax1.plot(y1)

fig.add_subplot(232)
fig.add_subplot(233)
fig.add_subplot(234)

ax2 = fig.add_subplot(235)
ax2.plot(y5)

ax3 = fig.add_subplot(236)
ax3.plot(y6)

plt.subplots_adjust(hspace=0.4, wspace=0.4)

plt.show()
```

##M2-4-4 設定子圖的座標範圍、座標說明文字與子圖標題

```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi)
y1 = np.sin(x)
y5 = np.cos(x)
y6 = np.tan(x)

fig = plt.figure(figsize=(8, 6))		#整個圖表大小為 8 x 6 英吋


ax1 = fig.add_subplot(231)
ax1.plot(x,y1)
ax1.set_title("y=sin(x)")
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
plt.xlim([0, 1])
plt.ylim([0, 1])

ax4 = fig.add_subplot(232)
ax4.set_xlabel('X-axis')
ax4.set_ylabel('Y-axis')

ax5 = fig.add_subplot(233)
ax5.set_xlabel('X-axis')
ax5.set_ylabel('Y-axis')

ax6 = fig.add_subplot(234)
ax6.set_xlabel('X-axis')
ax6.set_ylabel('Y-axis')

ax2 = fig.add_subplot(235)
ax2.plot(x,y5)
ax2.set_title("y=cos(x)")
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')
plt.xlim([0, 1])
plt.ylim([0, 1])

ax3 = fig.add_subplot(236)
ax3.plot(x,y6)
ax3.set_title("y=tan(x)")
ax3.set_xlabel('X-axis')
ax3.set_ylabel('Y-axis')
ax3.set_xlim([0, 1])
ax3.set_ylim([0, 1])

plt.subplots_adjust(hspace=0.4, wspace=0.4)

plt.show()
```

##M2-5-3 繪製堆疊長條圖

```
import numpy as np
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y1 = [12, 41, 32, 36, 21, 17]
y2 = [43, 1, 6, 17, 17, 9]

label1 = ['Apple', 'Orange', 'Banana', 'Pineapple', 'Kiwifruit', 'Stawberry']

plt.bar(x,y1,tick_label=label1,label="y1")
plt.bar(x,y2,tick_label=label1,label="y2",color='#FF8000',bottom=y1)
plt.legend(loc="upper right")

plt.show()
```

# M2-7 散佈圖 (test4)

## M2-7-1 散佈圖的繪製語法

```
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)

plt.scatter(x, y)
plt.show()
```

##M2-7-2 設定各資料點的樣式

```
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)

plt.scatter(x,y,marker='^',color='black')

plt.show()
```

##M2-7-3 設定資料點的大小

```
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)


plt.scatter(x,y,s=75)

plt.show()
```

##M2-7-4 給散佈圖的點套上不同深淺顏色

```
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)

size=[2*n for n in range(len(x))]
plt.scatter(x,y,c=size,s=size,cmap='spring')

plt.show()
```
